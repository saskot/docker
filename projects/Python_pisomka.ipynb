1. Definujte funkciu, ktorá bude mať 2 vstupy. A to maticu A a B. Táto funkcia Vám vráti súčet matíc A a B.

```python

def scitaj_matice(A, B):
    import numpy as np
    A_np = np.array(A)
    B_np = np.array(B)
    return A_np + B_np
```

4. Definujte funkciu, ktorá vráti menšie z 2 čísel ak su obe párne a väčšie ak čísla párne niesu.

```python
# Kod
a=5
b=10
def daco(a,b):
    if a % 2 == 0 and b % 2 == 0:
        return min(a, b)
    else:
        return max(a, b)
```
